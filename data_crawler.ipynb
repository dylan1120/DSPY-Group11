```python
def write_to_csv(existing_file, new_title, new_data):
    # 读取现有的CSV文件内容
    try:
        with open(existing_file, mode="r") as file:
            reader = csv.reader(file)
            existing_data = list(reader)
        # 如果文件不存在数据，则创建新的标题行
        if not existing_data:
            existing_data.append([])

        # 将新标题添加到标题行
        existing_data[0].append(new_title)

        # 将新数据添加到相应的列中
        for i, data in enumerate(new_data):
            if i < len(existing_data) - 1:
                existing_data[i + 1].append(data)
            else:
                existing_data.append([""] * (len(existing_data[0]) - 1) + [data])

        with open(existing_file, mode="w", newline="", encoding="utf-8") as file:
            writer = csv.writer(file)
            for item in existing_data:
                writer.writerow(item)
    except FileNotFoundError:
        with open(existing_file, mode="w", newline="", encoding="utf-8") as file:
            writer = csv.writer(file)
            writer.writerow([new_title])
            for item in new_data:
                writer.writerow([item])

    print(f"New data--{new_title} has been added to {existing_file}.")
```

```python
# 資料轉換與清理
def relative_date_to_absolute(relative_date_str):
    today = datetime.now()
    num, period = relative_date_str.split()[:2]

    if "天" in period:
        delta = relativedelta(days=int(num))
    elif "週" in period:
        delta = relativedelta(weeks=int(num))
    elif "月" in period:
        delta = relativedelta(months=int(num))
    elif "年" in period:
        delta = relativedelta(years=int(num))
    else:
        return None

    result_date = today - delta
    result_date_str = result_date.strftime("%Y-%m-%d")  # 格式化成 YYYY-MM-DD 的日期字串
    return result_date_str


df["dates"] = df["times"].apply(relative_date_to_absolute)
df["ratings"] = df["ratings"].str.extract("(\d+)").astype(int)
pattern = r"(在地嚮導)?\s*·\s*(\d+)\s*則評論\s*·\s*(\d+)\s*張相片"
df[["local_guide", "num_comments", "num_photos"]] = df["infs"].str.extract(pattern)
df["num_comments"] = (
    pd.to_numeric(df["num_comments"], errors="coerce").fillna(0).astype(int)
)
df["num_photos"] = (
    pd.to_numeric(df["num_photos"], errors="coerce").fillna(0).astype(int)
)

df = df.drop(columns=["times", "infs"])
df.to_csv("chiu_cleaned.csv", index=False)
print(df)
```
